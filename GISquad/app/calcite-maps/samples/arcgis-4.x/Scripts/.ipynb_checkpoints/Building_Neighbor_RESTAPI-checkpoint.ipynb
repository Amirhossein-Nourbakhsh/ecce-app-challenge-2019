```python
from arcgis import GIS
```

```python
from getpass import getpass
```

```python
import re, pandas as pd
```

```python
username = getpass("Enter username: ")
password = getpass("Enter password: ")
```

```python
gis = GIS("https://mcmaster.maps.arcgis.com/home/signin.html", username, password)
```

```python
parsed_items = gis.content.search("Parsed")
```

```python
parsed_items[1].layers[0].properties
```

```python
Xmin, Xmax, Ymin, Ymax, SpatialReference, neighborname, url = [],[],[],[],[],[],[]
for i in range(0,len(parsed_items[1].layers),1):
        Xmin.append(parsed_items[1].layers[i].properties.extent.xmin)
        Xmax.append(parsed_items[1].layers[i].properties.extent.xmax)
        Ymin.append(parsed_items[1].layers[i].properties.extent.ymin)
        Ymax.append(parsed_items[1].layers[i].properties.extent.ymax)
        SpatialReference.append(parsed_items[1].layers[i].properties.extent.spatialReference.wkid)
        neighborname.append(parsed_items[1].layers[i].properties.name)
        temp_url = str(parsed_items[1].layers[i])
        url.append(temp_url[re.search(":",temp_url).start()+1:-1])
```

```python
df = pd.DataFrame({"NeighborName":neighborname, "Url":url, 
                   "Xmin":Xmin, "Xmax":Xmax, 
                   "Ymin":Ymin, "Ymax":Ymax, 
                   "SpatialReference": SpatialReference})
```

```python
df.to_csv("../Data/Tabular/Neighbor_building_propertyInfo.csv", sep = ",", encoding = "utf-8")
```
